```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import numpy as np
import pandas as pd
import scipy
from scipy.optimize import curve_fit
plt.rc('xtick',labelsize=20); plt.rc('ytick',labelsize=20)
plt.rcParams['figure.figsize'] = (14,8)
```

I. Исследование пробега alfa-частиц с помощью счетчика Гейгера

```python
ax = plt.axes()
ax.xaxis.set_minor_locator(ticker.MultipleLocator(2));
ax.yaxis.set_minor_locator(ticker.MultipleLocator(1));

X=np.array([10,13,15,16,17,18,19,20,22,25,30,35,40])
N=np.array([15.0,15.4,15.8,15.2,9,2,0.5,0.4,0.3,0.3,0.2,0.2,0.2])

xerror=0.5
yerror=0.5

plt.errorbar(X,N, yerr=yerror, xerr=xerror, ecolor='black', elinewidth=1, capsize=2)

def model_f(x,a,b):
    return a*x + b

popt, pcov = curve_fit(model_f, X[2:8], N[2:8], p0=[1.4,26.8])

a_opt, b_opt = popt
x_model = np.linspace(min(X[2:8]), max(X[2:8]))
y_model = model_f(x_model, a_opt, b_opt)

plt.title('Зависимости скорости счета от расстояния', size=25)
plt.xlabel('X, мм',size=20)
plt.ylabel('N, c^-1',size=20)

txt = r" y = ax + b"
txt1 = r"a = -5.14 $\pm$ 0.34"
txt2 = r"b = 96.3 $\pm$ 10.4"
ax.text(33.8,15,txt, size = 20)
ax.text(34,14,txt1, size = 20)
ax.text(34,13,txt2, size = 20)

plt.scatter(X,N)
plt.plot(x_model,y_model, color='grey')
plt.grid()
```

II. Определение пробега alfa-частиц с помощью сцинтилляционного счетчика

```python
ax = plt.axes()
ax.xaxis.set_minor_locator(ticker.MultipleLocator(25));
ax.yaxis.set_minor_locator(ticker.MultipleLocator(400));

P=np.array([0,50,100,150,200,250,300,350,400,450,500,550,600,650,700,])
N=np.array([21998,20924,18960,16804,14596,13459,12466,12168,12068,12082,12277,12074,11982,11737,11924])

xerror=5
yerror=0

plt.errorbar(P,N, yerr=yerror, xerr=xerror, ecolor='black', elinewidth=1, capsize=2)

def model_f(x,a,b):
    return a*x + b

popt, pcov = curve_fit(model_f, P[:-8], N[:-8], p0=[1.4,26.8])

a_opt, b_opt = popt
x_model = np.linspace(min(P[:-8]), max(P[:-8]), 40)
y_model = model_f(x_model, a_opt, b_opt)

plt.title('Зависимости счета от давления',size=25)
plt.xlabel('P, Top',size=20)
plt.ylabel('N',size=20)

txt = r" y = ax + b"
txt1 = r"a = -2.14 $\pm$ 0.13"
txt2 = r"b = 52 $\pm$ 5.2"
ax.text(545,21000,txt, size = 20)
ax.text(550,20500,txt1, size = 20)
ax.text(550,20000,txt2, size = 20)


plt.scatter(P,N)
plt.plot(x_model,y_model, color='grey')
plt.grid()
```

III. Определение пробега alfa-частиц с помощью ионизационной камеры

```python
ax = plt.axes()
ax.xaxis.set_minor_locator(ticker.MultipleLocator(25));
ax.yaxis.set_minor_locator(ticker.MultipleLocator(50));

P=np.array([0,50,100,150,200,250,300,350,400,450,500,550,600,650,700,750])
I=np.array([9,52,127,205,290,376,460,555,653,752,856,940,953,944,930,917])

xerror=5
yerror=0

plt.errorbar(P,I, yerr=yerror, xerr=xerror, ecolor='black', elinewidth=1, capsize=2)

plt.axhline(y=948, color='grey', linestyle='--')
plt.axvline(x=600, color='grey', linestyle='--')

def model_f(x,a,b):
    return a*x + b

popt, pcov = curve_fit(model_f, P[:-3], I[:-3], p0=[1.4,26.8])

a_opt, b_opt = popt
x_model = np.linspace(min(P[:-3]), max(P[:-3]), 40)
y_model = model_f(x_model, a_opt, b_opt)

plt.title('Зависимости тока от давления',size=25)
plt.xlabel('P, Top',size=20)
plt.ylabel('I, пА',size=20)

txt = r" y = ax + b"
txtt= r"y = b"
txtt1= r" b = 940 $\pm$ 15"
txt1 = r"a = 1.69 $\pm$ 0.03"
txt2 = r"b = -49 $\pm$ 8"
ax.text(-5,850,txt, size = 20)
ax.text(0,790,txt1, size = 20)
ax.text(0,730,txt2, size = 20)
ax.text(0,600,txtt, size = 20)
ax.text(-6,540,txtt1, size = 20)

plt.scatter(P,I)
plt.plot(x_model,y_model, color='grey')
plt.grid()
```
